# 1 Create DataFrame

## 1.1 Create DataFrame from Array

```python
#Creating an array
data = np.array([[2,3],[4,5],[6,7]])
data
```

```python
#Creating DataFrame from array
pd.DataFrame(data,index=['row1','row2','row3',],columns=['col1','col2'])
```

## 1.2 Create DataFrame from Dictionary

```python
#Creating Lists
states = ['California','Florida','Texas','New York']
population = [23154354,23535,12343,643532]
```

```python
#Creating Dictionary
dict_states = {'States':states,'Population':population}
```

```python
#Creating DataFrame
df_population = pd.DataFrame(dict_states)
df_population
```

# Add a Column

```python
#Add new column
df_exams['language score']=70
df_exams.head()
```

```python
#using numpy, first create an array of 1000 items
language_score = np.arange(0,1000)
df_exams['language score'] = language_score
df_exams.head()
```

```python
#create random int values
language_score = np.random.randint(1,100,size=1000)
df_exams['language score'] = language_score
df_exams.head()
```

```python
#create random decimal values
language_score = np.random.uniform(1,100,size=1000)
df_exams['language score'] = language_score
df_exams.head()
```
